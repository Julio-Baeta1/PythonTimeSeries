```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import statsmodels.api as sm
from statsmodels.tsa.ar_model import AutoReg
#import statsmodels.formula.api as smf
```

```python
class simpleTimeSeries:
    
    def generateSeries(self,coeffs,const, initial_state, n, noise_var):
        
        data_series = np.zeros(n);
        white_noise = np.random.normal(0, noise_var, n)
        k = initial_state.shape[0];
        
        for i,ini_val in enumerate(initial_state):
            data_series[i] = ini_val
            #print(data_series[i])
        
        for i in range(k,n):
            data_series[i] = np.dot(coeffs, data_series[i-k:i]) + white_noise[i-1] +const
            #print(data_series[i-1])           
        #print(white_noise)
            
        return data_series
    
    def __init__(self, coeffs, const, initial_state, n, noise_var):
      self.time_series = self.generateSeries(coeffs,const,initial_state,n,noise_var)
    
    def printSeries(self):
        
        for iter in self.time_series:
            print(iter)
        
        #for val in initial_state:
        #    data_series[i] = val
        #    i += 1
            
        
```
